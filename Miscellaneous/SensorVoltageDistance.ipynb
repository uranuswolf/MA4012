```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
voltage = np.array([3.08, 0.97, 0.71, 0.57, 0.49,  0.42])
distance = np.array([4,    10,   15,  20,    25,   30])
```

```python
def sensor_model(V, A, B):
    return A / (V ** B)
```

```python
params, _ = curve_fit(sensor_model, voltage, distance)
A_fit, B_fit = params
print(f"Best Fit Formula: d = {A_fit:.2f} / V^{B_fit:.2f}")
```

```python
# Plot Data
plt.scatter(voltage, distance, label="Measured Data")
plt.plot(voltage, sensor_model(voltage, *params), label="Best Fit Curve", color="red")
plt.xlabel("Voltage (V)")
plt.ylabel("Distance (cm)")
plt.legend()
plt.show()
```
